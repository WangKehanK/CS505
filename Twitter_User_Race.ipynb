```python
usernames = df["Username"].values.tolist()
print(usernames)
len(usernames)
```

```python
auth_txt = "/content/drive/MyDrive/auth.txt"
f = open(auth_txt, "r")
keys = f.read()
f.close()

keys = keys.split('\n')
```

```python
# Get full names of each Twitter user using Tweepy
import tweepy
  
# assign the values accordingly
consumer_key = keys[0]
consumer_secret = keys[1]
access_token = keys[2]
access_token_secret = keys[3]
  
# authorization of consumer key and consumer secret
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
  
# set access to user's access key and access secret 
auth.set_access_token(access_token, access_token_secret)
  
# calling the api 
api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)

full_names = []
first_names = []
last_names = []

for screen_name in usernames:

  # fetching the user
  try:
    user = api.get_user(screen_name)
    
    # fetching the User's registered full name
    user_name = user.name

    # if user name contains first and last name, seperate them for prediction
    name = user_name.split(' ')
    # print(name)
    name_length = len(name)
    
    if name_length == 2:
      first_names.append(name[0])
      last_names.append(name[1])
    
    elif name_length == 3:
      first_names.append(name[0])
      last_names.append(name[2])

    # name not in proper first and last format
    else:
      full_names.append(user_name)

  except tweepy.error.TweepError:
    pass  
```

```python
# Create a new dataframe to store full names of twitter users
lastname_pred= pd.DataFrame()
fullname_pred= pd.DataFrame()

lastname_pred['Fullname'] = full_names
fullname_pred['Firstname'] = first_names
fullname_pred['Lastname'] = last_names
```

```python
# Predict Race using Last names of Twitter users
lastname_pred.head()
```

```python
census_prediction = pred_census_ln(lastname_pred, 'Fullname')
census_prediction[['Fullname', 'race']]
```

```python
wiki_prediction = pred_wiki_ln(lastname_pred, 'Fullname')
wiki_prediction[['Fullname', 'race']]
```

```python
fl_prediction = pred_fl_reg_ln(lastname_pred, 'Fullname')
fl_prediction[['Fullname', 'race']]
```

```python
# Predict Race using First and Last names of Twitter users
fullname_pred.head()
```

```python
wiki_full_pred = pred_wiki_name(fullname_pred, 'Firstname', 'Lastname')
wiki_full_pred[['Firstname', 'Lastname', 'race']]
```

```python
fl_full_pred = pred_fl_reg_name(fullname_pred, 'Firstname', 'Lastname')
fl_full_pred[['Firstname', 'Lastname', 'race']]
```

```python
nc_full_pred = pred_nc_reg_name(fullname_pred, 'Firstname', 'Lastname')
nc_full_pred[['Firstname', 'Lastname', 'race']]
```
